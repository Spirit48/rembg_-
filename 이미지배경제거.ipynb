```python
import time
import io
import numpy as np
import matplotlib.pyplot as plt

from PIL import Image, ImageFile

import rembg
from rembg import remove as remove_bg

ImageFile.LOAD_TRUNCATED_IMAGES = True

```

```python
image_path = "car.jpg"  # 이미지 파일의 경로를 지정해주세요.
image = Image.open(image_path)

plt.imshow(image)
plt.axis('off')  # 이미지 축을 표시하지 않도록 설정합니다.
plt.show()
```

```python
def remove_background(image_path):
    """
    Removes the background from an image using rembg and PIL.

    Args:
        image_path: 이미지 경로 입력

    Returns:
        배경이 제거된 이미지를 출력함
    """

    # 입력 이미지를 불러옴
    input_image = Image.open(image_path)

    # rembg를 사용하여 배경제거
    output_image = rembg.remove(input_image)

    # 출력 이미지 반환
    return output_image

# 실제 사용 예시
image_path = "car.jpg"
output_image = remove_background(image_path)

# 반환된 출력 이미지 화면
plt.imshow(output_image)
plt.axis('off')  # 이미지 축을 표시하지 않도록 설정합니다.
plt.show()
```
